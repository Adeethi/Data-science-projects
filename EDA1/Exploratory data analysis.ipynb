### 2. Contextual Overview
Fetal cardiotocography (CTG)  is a common method used to monitor fetal well-being during pregnancy and labor by assessing the fetal heart rate (FHR) and uterine activity.

#### Column-Specific Domain Knowledge
<b>1. LB (Baseline Fetal Heart Rate - FHR)</b><br>
Represents the average fetal heart rate, usually measured in beats per minute (BPM).
Normal baseline ranges are typically between 110-160 BPM. Abnormal values may indicate distress or other complications.

<b>2. AC (Accelerations)</b><br>
Accelerations are temporary increases in FHR and are considered a positive sign of fetal well-being.
They are often associated with fetal movement and suggest good oxygenation and a healthy autonomic nervous system.

<b>3. FM (Fetal Movements)</b><br>
Indicates the number or intensity of fetal movements detected.
Active fetal movement is a reassuring sign of well-being, while reduced movements may warrant further investigation.

<b>4. UC (Uterine Contractions)</b><br>
Refers to the frequency and intensity of uterine contractions.
Contractions can affect the FHR, and their interaction is crucial for assessing fetal well-being, especially during labor.

<b>5. DL (Decelerations Late)</b><br>
Late decelerations occur after the peak of a contraction and can indicate uteroplacental insufficiency (reduced blood flow/oxygen to the fetus).
They are a potential sign of fetal distress and need monitoring.

<b>6. DS (Decelerations Short)</b><br>
Short decelerations are brief dips in FHR that may not be concerning unless frequent or associated with other abnormal patterns.

<b>7. DP (Decelerations Prolonged)</b><br>
Prolonged decelerations are sustained drops in FHR that may indicate more severe fetal distress, such as cord compression or hypoxia.

<b>8. ASTV (Abnormal Short Term Variability)</b><br>
Refers to the percentage of time the short-term variability in FHR (beat-to-beat changes) is abnormal.
Reduced variability is often a concerning sign and may indicate compromised fetal health.

<b>9. MSTV (Mean Short Term Variability)</b><br>
Measures the average beat-to-beat variability in FHR.
Normal values typically range between 5-25 milliseconds. Variability outside this range may reflect fetal distress.

<b>10. ALTV (Abnormal Long Term Variability)</b><br>
Indicates the percentage of time the long-term variability in FHR (cyclic changes over minutes) is abnormal.
Abnormal patterns may suggest issues with fetal oxygenation or autonomic regulation.

<b>11. MLTV (Mean Long Term Variability)</b><br>
Refers to the average magnitude of cyclic changes in FHR over time.
Healthy patterns show a balance of variability, while abnormal values may warrant further investigation.

Earlier, in statistical summary we mentioned that AC, UC, FM and other veriability measures that have the possibility of error in data recording. 25th percentile of most of this records are 0 as well. So we are going to change those negative values to 0.

<b>Note: The emphasis here is on the values being negative rather than capping on the lower limit.</b>

```python
# List of columns to apply the transformation
columns_to_adjust = ["AC", "FM","UC","ASTV","MSTV","ALTV","MLTV"]

# Update values less than the 0 values
for col in columns_to_adjust:
    cleaner_fetal[col] = cleaner_fetal[col].apply(lambda x: max(x, 0))  # Replace values
cleaner_fetal.describe()
```

The above box plot confirms outliers in every column. We will also take a look at the histogram, before we decide on what to do with the outliers.

```python
# Adjust the figure size to fit all plots without scrolling
plt.figure(figsize=(12, 4 * 7))

# Plot histograms with KDE
for i, column in enumerate(cleaner_fetal.columns, 1):  # Start index from 1 for subplot numbering
    plt.subplot(7, 2, i)
    sns.histplot(cleaner_fetal[column], kde=True, bins=20, color='cornflowerblue', alpha=0.7)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Density')

# Adjust layout
plt.tight_layout()
plt.show()
```

From the histogram plots above, it is evident that none of the columns in the data table exhibit a symmetric distribution. The histograms display characteristics such as skewness, multimodality (multiple peaks), or heavy tails. Biological data is often "not normally distributed," and the graphs indicate that our dataset aligns with this observation.


At this point in the analysis, we have decided not to remove any outliers. This decision is based on several considerations:

<b>Expected Skewness and Outliers:</b> The skewness and presence of outliers are characteristic of cardiotocographic data, so their existence aligns with the nature of the dataset. Removing or altering them prematurely may distort the underlying trends and potentially erase meaningful information.

<b>Impact of Imputation or Capping:</b> Since there are numerous outliers, using methods such as imputation or capping could significantly alter the dataset, potentially leading to a loss of important insights or misrepresentation of the data's true behavior.

<b>Uncertainty of Outlier Origin:</b> At this stage, without clarity on their origin of the data, removing or modifying these points would be speculative and potentially harmful to the integrity of the analysis.

<b>Extreme Cases:</b> The extreme data points may correspond to critical cases such as fetal distress, making them natural outliers. These cases could provide important insights, so rather than removing them, the columns with such extreme data points can be flagged for separate analysis to better understand their implications.

And lastly, we need to learn more about the patterns in the data. We can decide to remove them later after initial analysis and additional input from domain experts.



<b>For NSP, Let's create a separate categorical column as it is expected to be categorical column. It will be helpful in analysis later. And do the same about the Tendency column, too</b>

```python
# Define bins and labels
bins = [0, 1.5, 2.5, 3.5]  # Adjust these based on your data
labels = ['Class 1', 'Class 2', 'Class 3']

# Use pd.cut to categorize
cleaner_fetal['NSP_cat'] = pd.cut(cleaner_fetal['NSP'], bins=bins, labels=labels, right=False)

# Verify the changes
cleaner_fetal.head(10)
```

```python
#replace null valuess with mode value
cleaner_fetal.loc[cleaner_fetal['NSP_cat'].isnull(), 'NSP_cat'] = 'Class 1'

cleaner_fetal.head(10)
```

```python
# Categorize Tendency into 'Negative', 'Neutral', and 'Positive'
def categorize_tendency(value):
    if value < 0:
        return 'Negative'
    elif value == 0:
        return 'Neutral'
    else:
        return 'Positive'

cleaner_fetal['Tendency_Category'] = cleaner_fetal['Tendency'].apply(categorize_tendency)
cleaner_fetal.head()
```

Note: We are using the scaled data to get a better view of the visualization

```python
# Calculate change between consecutive observations
scaled_fetal['FHR_change'] = scaled_fetal['LB'].diff()  # Difference between consecutive values
scaled_fetal['UC_change'] = scaled_fetal['UC'].diff()

# Plot the changes in FHR and UC
plt.figure(figsize=(15, 10))
plt.plot(scaled_fetal.index, scaled_fetal['FHR_change'], label="Change in FHR", linestyle='-', color='blue')
plt.plot(scaled_fetal.index, scaled_fetal['UC_change'], label="Change in UC", linestyle='--', color='red')
plt.title('Change in FHR and UC Over Consecutive Measurements')
plt.xlabel('Index')
plt.ylabel('Change in Value')
plt.legend()
plt.show()
```
