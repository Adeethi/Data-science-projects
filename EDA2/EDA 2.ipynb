## <h2 style="color:purple">2.  Feature Engineering </h2>

We will add some additional features into our feature set

1. Net Capital = capital_gain - capital_loss (captures overall financial gain/loss)
2. Age Group = Convert age into bins (e.g., Young (18-30), Middle-aged (31-50), Senior (51+))
3. Is Government Worker? = Convert workclass to 1 if "Federal-gov", "State-gov", "Local-gov", else 0
4. Work Intensity = Categorize hours_per_week (Part-time <35, Full-time 35-50, Overtime >50)

```python
#Net capital Feature
demographics['net_capital']  = demographics['capital_gain'] - demographics['capital_loss'] 
```

```python
# Age Group Feature
def categorize_age(age):
    if age <= 12:
        return 'Child'
    elif 12< age<=18:
        return 'Teenager'
    elif 18<age<=30:
        return 'Adult'
    elif 31 <= age <= 50:
        return 'Middle-aged'
    else:
        return 'Senior'

demographics['age_group'] = demographics['age'].apply(categorize_age)
```

```python
#Is Government Worker? Feature
gov_classes = ['Federal-gov', 'State-gov', 'Local-gov']
demographics['is_government_worker'] = demographics['workclass'].apply(lambda x: 1 if x in gov_classes else 0)
```

```python
#Work Intensity Feature
def categorize_work_hours(hours):
    if hours < 35:
        return 'Part-time'
    elif 35 <= hours <= 50:
        return 'Full-time'
    else:
        return 'Overtime'

demographics['work_intensity'] = demographics['hours_per_week'].apply(categorize_work_hours)
```

**Log transformation**

```python
# Select numerical features
numerical_features = ['age', 'final_weight', 'capital_gain', 'capital_loss', 'hours_per_week','net_capital']

# Check skewness
skewness = demographics[numerical_features].skew()
print(skewness)
```

We will log transform  the features that is highly skewed (where the skewness values is higher than 1)

```python
# Apply log transformation before scaling
demographics['capital_gain_log'] = np.log1p(demographics['capital_gain'])
demographics['capital_loss_log'] = np.log1p(demographics['capital_loss'])
demographics['fnl_wgt_log'] = np.log1p(demographics['final_weight'])
```

Also, all the negative values in the dataframe will be now zero because of the np.lop1p function. This is good because we dont expect our features to have negative values

## <h2 style="color:purple">4. Feature Selection</h2>

**Outlier detection through Isolation Forest Technique**

```python
demographics_encoded.info()
```

```python
# Select numerical features for outlier detection
num_cols = ['age', 'final_weight', 'capital_gain', 'capital_loss', 'hours_per_week','net_capital','capital_gain_log', 'capital_loss_log']
df_num = demographics_encoded[num_cols]
```

```python
df_num.isna().sum()
```

```python
from sklearn.ensemble import IsolationForest

# Initialize Isolation Forest model
iso_forest = IsolationForest(contamination=0.03, random_state=42)  # Adjust contamination if needed
```

Ideally we put contamination rate between 1 to 5%

```python
# Fit and predict outliers
demographics_encoded["outlier"] = iso_forest.fit_predict(df_num)

# Remove outliers (keeping only non-outliers)
df_cleaned = demographics_encoded[demographics_encoded["outlier"] == 1].drop(columns=["outlier"])

# Display how many outliers were removed
print(f"Outliers detected and removed: {demographics_encoded.shape[0] - df_cleaned.shape[0]}")
```

```python
df_cleaned.shape
```

**How Outliers Affect Model Performance**

Outliers can significantly impact a model's performance by:
- Distorting Mean & Variance: Can lead to biased parameter estimation.
-  Affecting Model Accuracy: Outliers can introduce noise, making the model less generalizable.
-  Impacting Feature Scaling: Can cause improper normalization or skewed distributions.
- Reducing Model Stability: Especially in tree-based and regression models.

#### <h3 style="color:midnightblue">Scaling</h3>

1. Standard Scaler
- Preferred in: Machine learning models like Logistic Regression, SVM, PCA (sensitive to variance).

2. Min Max Scaler
- Preferred in: Neural networks & deep learning (activations like ReLU work better in a fixed range).
- When you want to preserve relationships between variables or when values need to be in a bounded interval


For our case, as the dataset set is used for prediction, we will use Standard Scaler
